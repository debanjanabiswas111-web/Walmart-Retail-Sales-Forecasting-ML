```python
# Import libraries
import pandas as pd
import numpy as np
```

```python
#Load dataset
df = pd.read_csv('Walmart.csv')
```

```python
#Display basic info
df.head()
```

```python
#Check missing values
df.isna().sum()
```

```python
#Handling missing values

df= df.ffill()
```

```python
#Convert Date Column
df["Date"] = pd.to_datetime(df["Date"], format="%d-%m-%Y")


#Feature Engineering
df['Year'] = df['Date'].dt.year
df['Month'] = df['Date'].dt.month
df["Week"] = df["Date"].dt.isocalendar().week # .week is deprecated, use isocalendar().week ##to isolate just the speciic week column
```

```python
df.to_csv("Data_cleaning.csv", index=False)

print("Cleaned data saved successfully.")
```
